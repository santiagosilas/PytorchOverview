<font color='0471A6'><h2>⭐ Funções de Ativação em Pytorch</h2></font>

```python
import torch
import matplotlib.pyplot as plt

xs = torch.linspace(-4, 4, 100)
ys = torch.relu(xs)

plt.plot(xs.numpy(), ys.numpy(), label='relu')

ys = torch.sigmoid(xs)
plt.plot(xs.numpy(), ys.numpy(),label='sigmoid')

ys = torch.tanh(xs)
plt.plot(xs.numpy(), ys.numpy(), label='tanh')
plt.grid()
plt.legend()
```
